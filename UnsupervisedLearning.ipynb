```python
#k mean clustering #first time randomly assign centren
li = [110,2,5,20,40,50,100,150]
cc1 = 7
cc2 = 45
cluster1 = []
cluster2 = []
for i in li:
  if(abs(i-cc1)<abs(i-cc2)):
    cluster1.append(i)
  else:
    cluster2.append(i)
print(cluster1, cluster2)
```

```python
#caluculating centre of cluster #second iteration
cc1 = sum(cluster1)/len(cluster1)
cc2 = sum(cluster2)/len(cluster2)
cluster1 = []
cluster2 = []
for i in li:
  if(abs(i-cc1)<abs(i-cc2)):
    cluster1.append(i)
  else:
    cluster2.append(i)
print(cluster1, cluster2)
```

```python
#caluculating centre of cluster #third iteration
cc1 = sum(cluster1)/len(cluster1)
cc2 = sum(cluster2)/len(cluster2)
cluster1 = []
cluster2 = []
for i in li:
  if(abs(i-cc1)<abs(i-cc2)):
    cluster1.append(i)
  else:
    cluster2.append(i)
print(cluster1, cluster2)
```
